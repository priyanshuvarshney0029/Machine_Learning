```python
import numpy as np
import pandas as pd
import seaborn as sns

np.random.seed(37)

n = 10
labels = [f'x{i}' for i in range(n)]
data = pd.DataFrame(np.random.randint(0, 11, size=(n, n)), index=labels, columns=labels)

fig, ax = plt.subplots(figsize=(8, 8))

_ = sns.heatmap(data)
_ = ax.set_title('Basic heatmap plot')

y_min, y_max = ax.get_ylim()
dh = 0.8
_ = ax.set_ylim(y_min + dh, y_max - dh)
```

```python
y=data['MEDV']
data.drop(columns=['MEDV'],inplace=True)
x=data
```

```python
data.plot(kind='box')
```

```python
plt.scatter(x['ZN'],y,color='blue',label='ZN')

plt.plot(x['ZN'],y,color='red',label='MEDV',linewidth=2)

plt.xlabel('Feature (x[ZN]')
plt.ylabel('Target (y[MEDV]')
plt.title('ZN vs MEDV')
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("Mean squared Error:",mse)
print("R2 Score:",r2)
```
